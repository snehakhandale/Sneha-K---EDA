## write a function which will take dictionary of crew column and return me the list of producers and if there no producers then it will give me []

```python
sample_dict=df2_crew['crew'].iloc[1]
```

```python
df2_crew
```

```python
sample_dict
```

```python
list_producers=[]

for i in sample_dict:
    if i['job']=='Producer':
        list_producers.append(i['name'])
```

```python
list_producers
```

```python
def producers_list(dict_prod):
    list_producers=[]

    for i in dict_prod:
        if i['job']=='Producer':
            list_producers.append(i['name'])

    return list_producers
```

```python
producers_list(df2_crew['crew'])
```

```python
you need to make a column where for every film list of producers will be stored
```

```python
df2_crew['producers_list']=df2_crew['crew'].apply(producers_list)
```

write a function which will take dictionary of crew column and return me the list of producers and if there no producers then it will give me []

```python
def directors_list(dict_prod):
    list_directors=[]

    for i in dict_prod:
        if i['job']=='Director':
            list_directors.append(i['name'])

    return list_directors
```

```python
df2_crew['directors_list']=df2_crew['crew'].apply(directors_list)
```

```python
df2_crew1=df2_crew[['title','producers_list','directors_list','roi']]
```

```python
##show top 3 movies having highest ROI.

df2_crew_roi = df2_crew1[df2_crew1['roi']!=np.inf]
```

```python
df2_crew_roi.sort_values('roi',ascending=False).reset_index(drop=True).iloc[2:5]
```

Q5. Which actor has acted in the most number of movies? Deep dive into the movies, genres and profits corresponding to this actor.

```python
df2_cast=df1[df1['cast'].notna()].reset_index(drop=True)
```
